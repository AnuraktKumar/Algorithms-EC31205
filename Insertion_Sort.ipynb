<h3> Generating a sequence of integers <\h3>

```python
a0 = list(range(100))
a1 = list(range(500))
a2 = list(range(1000))
a3 = list(range(5000))
a4 = list(range(10000))
a5 = list(range(15000))
a6 = list(range(30000))
length_arr = [len(a0),len(a1),len(a2),len(a3),len(a4),len(a5),len(a6)]
theoretical_runntime = [l*l for l in length_arr]
```

<h3> Insertion Sort with Runtime = $O(n^2)$<\h3>

```python
A = [5,2,1,3]
def insertion_sort(A, n):
    #Base Case
    if(n<=1):
        return
    
    #Sorting the n-1 elements first
    insertion_sort(A, n-1)
    
    #Inserting the last elements into its correct postion
    last = A[n-1]
    j = n-2
    while(j>=0 and A[j]>last):
        A[j+1] = A[j]
        j = j-1
    A[j+1] = last
```

<h3> Using the time library to get the actual runntime of the recursive insertion sort algorithm <\h3>

```python
import time
timeArr = []
#On array a0
startTime = time.time()
insertion_sort(a0, len(a0))
timeArr.append(time.time() - startTime)
#On array a1
startTime = time.time()
insertion_sort(a1, len(a1))
timeArr.append(time.time() - startTime)
#On array a2
startTime = time.time()
insertion_sort(a2, len(a2))
timeArr.append(time.time() - startTime)
#On array a3
startTime = time.time()
insertion_sort(a3, len(a3))
timeArr.append(time.time() - startTime)
#On array a4
startTime = time.time()
insertion_sort(a4, len(a4))
timeArr.append(time.time() - startTime)
#On array a5
startTime = time.time()
insertion_sort(a5, len(a5))
timeArr.append(time.time() - startTime)
#On array a6
startTime = time.time()
insertion_sort(a6, len(a6))
timeArr.append(time.time() - startTime)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2,1,figsize=[12,8])
#Plotting the actual runntime
ax[0].set_title("Actual Runtime of Recursive Insertion Sort")
ax[0].set_xlabel("Size of the array --->")
ax[0].set_ylabel("Runntime --->")
ax[0].plot(length_arr, timeArr, color='red',label="Actual Runntime")
ax[0].legend()
#Plotting the theoretical runntim
ax[1].set_title(" TheoreticalRuntime of Recursive Insertion Sort")
ax[1].set_xlabel("Size of the array --->")
ax[1].set_ylabel("Runntime --->")
ax[1].plot(length_arr, theoretical_runntime, color='orange', label="Theoretical Runntime")
ax[1].legend()
fig.tight_layout()
```

The following observations are made
<p>
<ol>
<li>The actual runtime of the algorithm turns out to the $T = \frac{n(n+1)}{2}$, but in the big Oh notation it is $O(n^2)$. <\li>
<li>The actual runtime curve therefore follows the $T = n^2$ curve but only approximately, the actual time curve reaches closer as the length of array increases. <\li>
<\ol>
