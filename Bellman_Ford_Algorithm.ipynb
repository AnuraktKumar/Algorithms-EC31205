<h2> Implementation of the Bellman Ford's Algorithm <\h2>

```python
class Edge:
    def __init__(self, src, trgt, wt):
        self.a = src
        self.b = trgt
        self.cost = wt

class Graph:
    def __init__(self, numNodes, numEdges):
        self.n = numNodes
        self.m = numEdges
        self.edges = []
    def addEdge(self, a, b, l):
        e = Edge(a-1,b-1,l)
        self.edges.append(e)
        print("Added an edge from Node", a, "to Node", b, "of length:", l)
    def printDistAndPath(self, s, minDistances, p):
        for i in range(len(minDistances)):
            currNode = i
            print("Distance between Node", s+1, "and Node", currNode+1, "is:", minDistances[i], end=' | Path: ')
            while(currNode!=s):
                print(currNode+1, "<-", end=" ")
                currNode = p[currNode]
            print(s+1)
    def runBellmanFord(self, s):
        #Declaring the source node, the max. dist, bringing the edge array, number of nodes and number of edges
        s = s-1
        INF = 100000000
        edges = self.edges
        n = self.n
        m = self.m
        #Declaring the distance and predecessor array to store the minimum distances and the predecessors
        d = [INF]*n
        p = [0]*n
        d[s] = 0
        #For v-1 times, for each edge if the distance 
        for i in range(n-1):
            for j in range(m):
                if(d[edges[j].a]<INF):
                    d[edges[j].b] = min(d[edges[j].b], d[edges[j].a] + edges[j].cost)
                    p[edges[j].b] = edges[j].a
        self.printDistAndPath(s,d,p)
        return d
```
